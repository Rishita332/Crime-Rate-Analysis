```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv('segregated_monthly_data.csv')
x = crime.Year
y = crime.Month
z = crime.Count
data = pd.DataFrame({'Year': x, 'Month': y, 'Crime incidents': z})
data_pivoted = data.pivot("Year", "Month", "Crime incidents")
ax = sns.heatmap(data_pivoted, linewidth=0.5)
plt.show()
```
